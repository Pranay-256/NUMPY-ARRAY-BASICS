```python
#numpy indexing and slicing of matrix
```

```python
arr = np.arange(1,31).reshape(6,5)
arr
```

```python
arr[0]#indexing of rows
```

```python
arr[5]
```

```python
arr[0,0]#print element at 0th row and 0th column
```

```python
arr[0,4]
```

```python
arr[5,4]
```

```python
#array slicing
```

```python
slice=arr[0:2,1:3] #0 to 1st row end -1 , 1 to 2nd column end-1
slice
```

```python
#slice2 = arr[3:6,3:5]
slice2 = arr[3:,3:]
slice2
```
